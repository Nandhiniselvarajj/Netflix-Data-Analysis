```python
import pandas as pd  # This helps us work with data
import matplotlib.pyplot as plt  # This helps us draw charts
import seaborn as sns  # This helps make beautiful charts

# This line tells the notebook to show charts inside the page
%matplotlib inline
```

```python
# Read the dataset
df = pd.read_csv('netflix_titles.csv')

# Show the first 5 rows of the table
df.head()
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Drop rows where title or date_added is missing
df = df.dropna(subset=['title', 'date_added']).copy()

# Fill missing values in relevant columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')

# Strip leading/trailing spaces from date column (fix the formatting issue)
df['date_added'] = df['date_added'].str.strip()

# Convert to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Drop rows where conversion failed (if any)
df = df.dropna(subset=['date_added']).copy()

# Add new date columns
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()

# Check that all missing values are gone
df.isnull().sum()
```

```python
# Count the number of Movies and TV Shows
type_counts = df['type'].value_counts()

# Plot a bar chart
type_counts.plot(kind='bar', color=['salmon', 'skyblue'])
plt.title('Number of Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
# Count how many shows/movies were added each year
yearly_data = df['year_added'].value_counts().sort_index()

# Plot a line chart
plt.figure(figsize=(10, 5))
yearly_data.plot(kind='line', marker='o', color='green')
plt.title('Netflix Content Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
# Split the genre strings into individual genres
all_genres = df['listed_in'].str.split(', ')

# Flatten the list of lists into one big list
from collections import Counter
genre_list = sum(all_genres, [])

# Count the frequency of each genre
genre_counts = Counter(genre_list)

# Turn it into a DataFrame and plot top 10
genre_df = pd.DataFrame(genre_counts.most_common(10), columns=['Genre', 'Count'])

# Plot it
# Improved bar plot with centered labels
# Plot without future warning
plt.figure(figsize=(10, 5))
sns.barplot(data=genre_df, x='Genre', y='Count', hue='Genre', palette='magma', legend=False)

plt.title('Top 10 Most Common Genres on Netflix')
plt.xticks(rotation=70, ha='center')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.grid(axis='y')
plt.show()


```
